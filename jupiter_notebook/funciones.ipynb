## funciones lambda (funaiones anómimas)

```python
suma = lambda x, y: x + y #función amonima
print(suma(2,89))
```

```python
cuadrados = lambda x: x **2
print(cuadrados(4))
```

## funciones con argumentos

```python
def aplicar(funcion, valor):
    return funcion(valor)
```

```python
print(aplicar(lambda x:x*2,5)) #multiplica por 2 el valor 5
```

```python
def cuadrado(x):
    return x **2
```

```python
print(aplicar(cuadrado, 9))
```

```python
def imprimir_despues(func): #función que recibe otra función como argumento
    print("Antes de la función")
    func() #llama a la función pasada como argumento
    print("despues de la función")
```

```python
imprimir_despues(lambda:print("función ajecutada"))
```

## Argunanto arbitrarios
- *args agrupa múltiples argumentos posicionales en una tupla
- **kwargs agrupa mútiples argumentos nombrados en un diccionario

```python
def lista_elementos(*args):
    for elementos in args: #itera sobre los elementos recibidos
        print(f"-{elementos}")
```

```python
lista_elementos(" manzana", " banana", " cereza")
```

```python
def sumar_num(*numeros):
    return sum(numeros)
```

```python
print(sumar_num(5,10,15,20,56))
```

```python
def mostrar_info(**kwargs): 
    for clave, valor in kwargs.items(): #itera sobre los elementos del diccionario
        print(f"{clave}: {valor}") #imprime clave y valor
```

```python
mostrar_info(nombre="jhon", edad=30, ciudad="New York")
```

```python
def construir_perfil(**datos):
    return datos
```

```python
prefil = construir_perfil(nombre="jhon", edad=30, ciudad="New York") #almacena el dicionario retornado en una variable
print(prefil)
```

```python
def imprimir_conifg(**config):
    for clave in config:
        print(f"{clave.upper()}: {config[clave]}")
```

```python
imprimir_conifg(resolucion="1025p", fps=60,formato="mp4")
```
