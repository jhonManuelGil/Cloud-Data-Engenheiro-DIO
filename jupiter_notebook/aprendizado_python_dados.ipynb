##  Data-Driven Python



```python
a = 10
b = 3

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
modulo = a % b
potencia = a ** b
divisao_inteira = a // b

print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
print("Módulo:", modulo)
print("Potência:", potencia)
print("Divisão Inteira:", divisao_inteira)
```

```python
x = True
y = False

resultado_and = x and y
resultado_or = x or y
resultado_not = not x

print("Resultado do 'e lógico':", resultado_and)
print("Resultado do 'ou lógico':", resultado_or)
print("Resultado da negação lógica:", resultado_not)
```

```python
a = 10
b = 12

print(f"a == b : {a == b}")
print(f"a != b : {a != b}")
print(f"a > b : {a > b}")
print(f"a < b : {a < b}")
print(f"a >= b : {a >= b}")
print(f"a <= b : {a <= b}")
print(f"a > 5 and b > 5 : {a > 5 and b > 5}")
print(f"a > 5 or b < 5 : {a > 5 or b < 5}")
print(f"not(a > b) : {not(a > b)}")
```

```python
numero1 = 10
numero2 = 20

igual = numero1 == numero2
diferente = numero1 != numero2
maior_que = numero1 > numero2
menor_que = numero1 < numero2
maior_ou_igual = numero1 >= numero2
menor_ou_igual = numero1 <= numero2

print("Igual a:", igual)
print("Diferente de:", diferente)
print("Maior que:", maior_que)
print("Menor que:", menor_que)
print("Maior ou igual a:", maior_ou_igual)
print("Menor ou igual a:", menor_ou_igual)
```

```python
nomes = ["Alice", "Bob", "Charlie", True]

print(nomes)

#lista de lidta

lsita_lista = [ [1, 2, 3], ['a', 'b', 'c'], [True, False], [1.5, 2.5], ['x', 'y', 'z'], [10, 20, 30] ]

print(lsita_lista[0]) # Acessa a primeira sublista
print(lsita_lista[4])
print(lsita_lista[4][-2]) # Acessa o segundo elemento da quinta sublista

# adicionando itens a uma lista append()
frutas = ["maçã", "banana", "laranja"]
frutas.append("uva")
print(frutas)  # Saída: ['maçã', 'banana', 'laranja', 'uva']
frutas.pop(0)
print(frutas)  # Saída: ['banana', 'laranja', 'uva']
print(frutas.pop(0))  # Saída: 'maçã'
# remove() remove o último item da lista se nenhum índice for especificado
frutas.remove(frutas[0])
print(frutas)  # Saída: ['uva']
del frutas[0]
print(frutas)
nomes.append("jhon")
nomes.append("Adrian")
nomes.append("Roberth")
nomes.append("Daniel")
nomes.append("Duclas")
print(nomes)


```
