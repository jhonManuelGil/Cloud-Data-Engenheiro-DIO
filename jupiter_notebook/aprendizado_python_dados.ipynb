##  Data-Driven Python



```python
a = 10
b = 12

print(f"a == b : {a == b}")
print(f"a != b : {a != b}")
print(f"a > b : {a > b}")
print(f"a < b : {a < b}")
print(f"a >= b : {a >= b}")
print(f"a <= b : {a <= b}")
print(f"a > 5 and b > 5 : {a > 5 and b > 5}")
print(f"a > 5 or b < 5 : {a > 5 or b < 5}")
print(f"not(a > b) : {not(a > b)}")
```

```python
age = 30

if age < 18:
    print("Menor de idade")
elif age >= 18: 
    print("Maior de idade")
else:
 print(f"edade não encontarda")
```
