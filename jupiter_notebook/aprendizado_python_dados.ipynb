##  Data-Driven Python



```python
a = 10
b = 3

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
modulo = a % b
potencia = a ** b
divisao_inteira = a // b

print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
print("Módulo:", modulo)
print("Potência:", potencia)
print("Divisão Inteira:", divisao_inteira)
```

```python
x = True
y = False

resultado_and = x and y
resultado_or = x or y
resultado_not = not x

print("Resultado do 'e lógico':", resultado_and)
print("Resultado do 'ou lógico':", resultado_or)
print("Resultado da negação lógica:", resultado_not)
```

```python
a = 10
b = 12

print(f"a == b : {a == b}")
print(f"a != b : {a != b}")
print(f"a > b : {a > b}")
print(f"a < b : {a < b}")
print(f"a >= b : {a >= b}")
print(f"a <= b : {a <= b}")
print(f"a > 5 and b > 5 : {a > 5 and b > 5}")
print(f"a > 5 or b < 5 : {a > 5 or b < 5}")
print(f"not(a > b) : {not(a > b)}")
```

```python
numero1 = 10
numero2 = 20

igual = numero1 == numero2
diferente = numero1 != numero2
maior_que = numero1 > numero2
menor_que = numero1 < numero2
maior_ou_igual = numero1 >= numero2
menor_ou_igual = numero1 <= numero2

print("Igual a:", igual)
print("Diferente de:", diferente)
print("Maior que:", maior_que)
print("Menor que:", menor_que)
print("Maior ou igual a:", maior_ou_igual)
print("Menor ou igual a:", menor_ou_igual)
```

```python
# admin, user, guest estrutura de controle de acesso if elif else

rule = input("Digite o tipo de usuário (admin, user, guest): ")

if rule == "admin":
    print("Acesso total concedido.")
elif rule == "user":
    print("Acesso limitado concedido.")
elif rule == "guest":
    print("Acesso de convidado concedido.")
else:
    print("Acesso restrito.")
```
